```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
import pandas as pd
data = [0, 100, 120, 130, 150, 180, 200, 220, 250, 400]
df = pd.DataFrame(data, columns=['Values'])
Q1 = df['Values'].quantile(0.25)
Q2 = df['Values'].quantile(0.50)
Q3 = df['Values'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Values'] < lower_bound) | (df['Values'] > upper_bound)]
print(f"Q1 (25th percentile): {Q1}")
print(f"Q2 (Median / 50th percentile): {Q2}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Q3 - Q1): {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print("\nOutliers:")
print(outliers['Values'].values)

```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Values'], color='lightgreen')

Q1 = df['Values'].quantile(0.25)
Q2 = df['Values'].median()
Q3 = df['Values'].quantile(0.75)
plt.axvline(Q1, color='blue', linestyle='--', label='Q1')
plt.axvline(Q2, color='black', linestyle='-', label='Median (Q2)')
plt.axvline(Q3, color='blue', linestyle='--', label='Q3')


IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Values'] < lower_bound) | (df['Values'] > upper_bound)]
for outlier in outliers['Values']:
    plt.plot(outlier, 0, 'ro', label='Outlier' if outlier == outliers['Values'].values[0] else "")


plt.title("Box Plot with Q1, Q2, Q3 and Outlier")
plt.legend()
plt.grid(True)
plt.show()
```
